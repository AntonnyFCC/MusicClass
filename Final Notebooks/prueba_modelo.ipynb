# Prueba del modelo con nueva canción

```python
import librosa
import cv2 as cv
import numpy as np

import IPython
```

```python
def get_mel_spectogram3(onda):
    mewSpectogram = librosa.feature.melspectrogram(y=onda, sr=22050 , hop_length=512)
    mel_spectrogram_log = librosa.power_to_db(mewSpectogram, ref=np.max)
    mel_spectrogram_log = mel_spectrogram_log.astype(np.uint8)
    spec_img = cv.applyColorMap(mel_spectrogram_log, cv.COLORMAP_JET)
    spec_img = cv.cvtColor(spec_img, cv.COLOR_BGR2RGB)
    return spec_img
```

```python
import tensorflow as tf
```

```python
modelo = tf.keras.models.load_model('Modelos/ResCorcinoNetV100.h5')
```

# Función para predecir

```python
def make_predic(image_2_use):
    imagen_redimensionada = cv.resize(image_2_use, (1293, 128))
    sample = np.expand_dims(imagen_redimensionada, axis=0)

    prediccion = modelo.predict(sample)
    clase_predicha = np.argmax(prediccion)

    clases = ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']

    return 'El género de la música es ' + clases[clase_predicha]
```

## Con audios de prueba


```python
IPython.display.Audio('Audios prueba\Classical\la-capricciosa-30-sec-110908.mp3')
```

```python
wave, sam_r = librosa.load('Audios prueba\Classical\la-capricciosa-30-sec-110908.mp3',sr=22050)
```

```python
img_spec = get_mel_spectogram3(wave)
```

```python
make_predic(img_spec)
```
